## 3. ***Data Wrangling***

#### Chart - 10: Runtime vs IMDb Score- Bivariate

```python
# Chart - 10 visualization code
plt.scatter(titles['runtime'], titles['imdb_score'])

plt.title("Runtime vs IMDb Score")
plt.xlabel("Runtime")
plt.ylabel("IMDb Score")

plt.show()
```

##### What is the insight found from the chart?

1. Most titles have runtime between 50 to 150 minutes
2. IMDb scores mostly range between 5 and 8.
3. The points are widely scattered vertically, which indicates that longer runtime does not necessarily lead to higher IMDb scores.
4. Short titles can have high ratings and Long titles can have low ratings.

#### Chart - 11: Correlation Heatmap- multivariate

```python
# Chart - 11 visualization code
numeric_cols = titles[['runtime','imdb_score','imdb_votes']]

corr = numeric_cols.corr()

sns.heatmap(corr, annot=True)

plt.title("Correlation Heatmap")

plt.show()
```

##### 1. Why did you pick the specific chart?

It's a multivariate chart which shows the relationship between various properties.

##### 2. What is the insighs found from the chart?


The correlation heatmap shows very weak relationships between runtime, IMDb score, and IMDb votes.
Runtime has almost no correlation with IMDb score (-0.09), indicating that longer titles do not necessarily receive better ratings.
IMDb score and IMDb votes show a weak positive correlation (0.16), suggesting that highly rated titles tend to receive slightly more votes.
Overall, the variables do not show strong dependencies.
